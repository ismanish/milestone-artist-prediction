```python
import pandas as pd
from google.cloud import storage
import os
```

```python
local_dir = "./data"
bucket_name = "wikiart_data_sample"
path_to_json = "ms_personal_gmail.json"
```

```python
#####Final Code
```

```python
def download_data_local(local_dir, bucket_name, path_to_json = False):
    
    if path_to_json:
        storage_client = storage.Client.from_service_account_json(path_to_json)
    else:
        storage_client = storage.Client()
        
    bucket = storage_client.get_bucket(bucket_name)
    blobs = bucket.list_blobs()
    
    for blob in blobs:
        file_path = os.path.join(local_dir, blob.name)
        os.makedirs(os.path.dirname(file_path),exist_ok = True)
        blob.download_to_filename(file_path)
```

```python
download_data_local(local_dir, bucket_name)
```
